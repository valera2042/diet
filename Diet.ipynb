# Data Cleaning

Plan for cleaning

1. Some columns like  'Associate/technical degree' and 'Bachelor degree or higher' should be merged into one.
2. Education : remove words degree or higher.
3. Points: data in brakets must be deleted.
4. Income must have only Low, middle and High words, create a separate column for the salary values.
5. Remove values in brakets from income amount, L, food factor and goal.
6. Prepare columns: there shoul not be a white space between words in columns' names.

1. Some columns like 'Associate/technical degree' and 'Bachelor degree or higher' should be merged into one.

```python
# melt bachelor and associate degree

def melt_bachelor_associate(df_concat):
    
    # columns that will not be changed
    no_change = [        'Low ($0-$49,999)',        'Middle ($50,000-$99,999)',
                          'High ($100,000+)',    'Not living with children/SOd',
               'Living with SO, no children',     'Living with SO and children',
               'Living with children, no SO', 'Does not prepare food regularly',
           'Prepares food at home regularly',                           
                'Not overweight (<25 kg/m2)',     'Overweight (25-29.99 kg/m2)',
                       'Obesity (?30 kg/m2)',           'Nothing in particular',
                               'Lose weight',                     'Gain weight',
                      'Stay the same weight']
    
    # columns that will be changed
    to_change = ['Associate/technical degree', 'Bachelor degree or higher']
    
    # performing melting
    df_concat = pd.melt(df_concat, id_vars=no_change, value_vars=to_change,
            var_name='Education', value_name='Points')
    
    return df_concat

df_concat = melt_bachelor_associate(df_concat)
```

```python
# working with income

def income(df_concat):
    
    # columns that will not be changed
    no_change = [        'Not living with children/SOd', 'Living with SO, no children',
           'Living with SO and children', 'Living with children, no SO',
           'Does not prepare food regularly', 'Prepares food at home regularly',
           'Not overweight (<25 kg/m2)', 'Overweight (25-29.99 kg/m2)',
           'Obesity (?30 kg/m2)', 'Nothing in particular', 'Lose weight',
           'Gain weight', 'Stay the same weight', 'Education', 'Points']

    # columns that will be changed
    to_change = ['Low ($0-$49,999)', 'Middle ($50,000-$99,999)', 'High ($100,000+)']

    # performing melting
    df_concat = pd.melt(df_concat, id_vars=no_change, value_vars=to_change,
            var_name='Income', value_name='Income_amount')

    return df_concat

df_concat = income(df_concat)
```

```python
# living situation

def living_situation(df_concat):
    
    # columns that will not be changed
    no_change = [       'Does not prepare food regularly', 'Prepares food at home regularly',
           'Not overweight (<25 kg/m2)', 'Overweight (25-29.99 kg/m2)',
           'Obesity (?30 kg/m2)', 'Nothing in particular', 'Lose weight',
           'Gain weight', 'Stay the same weight', 'Education', 'Points', 'Income',
           'Income_amount']

    # columns that will be changed
    to_change = ['Not living with children/SOd', 'Living with SO, no children',
           'Living with SO and children', 'Living with children, no SO']

    # performing melting
    df_concat = pd.melt(df_concat, id_vars=no_change, value_vars=to_change,
            var_name='Living Situation', value_name='Living points')

    return df_concat

df_concat = living_situation(df_concat)
```

```python
# foof preparation

def food(df_concat):
    
    # columns that will not be changed
    no_change =  [             'Not overweight (<25 kg/m2)', 'Overweight (25-29.99 kg/m2)',
           'Obesity (?30 kg/m2)', 'Nothing in particular', 'Lose weight',
           'Gain weight', 'Stay the same weight', 'Education', 'Points', 'Income',
           'Income_amount', 'Living Situation', 'Living points']
    # columns that will be changed
    to_change = ['Does not prepare food regularly', 'Prepares food at home regularly']
    # performing melting
    df_concat = pd.melt(df_concat, id_vars=no_change, value_vars=to_change,
            var_name='Food preparation', value_name='Food factor')
    
    return df_concat

df_concat = food(df_concat)
```

```python
# Obesity indicator

def food(df_concat):
    
    # columns that will not be changed
    no_change =  [ 'Nothing in particular', 'Lose weight',
           'Gain weight', 'Stay the same weight', 'Education', 'Points', 'Income',
           'Income_amount', 'Living Situation', 'Living points', 'Food preparation',
           'Food factor']

    # columns that will be changed
    to_change = ['Not overweight (<25 kg/m2)', 'Overweight (25-29.99 kg/m2)',
           'Obesity (?30 kg/m2)']
    # performing melting
    df_concat = pd.melt(df_concat, id_vars=no_change, value_vars=to_change,
            var_name='Obesity indicator', value_name='Obesity factor')
    
    return df_concat

df_concat = food(df_concat)
```

```python
# What is your goal

def goal(df_concat):
    
    # columns that will not be changed
    no_change =  ['Education', 'Points', 'Income',
           'Income_amount', 'Living Situation', 'Living points', 'Food preparation',
           'Food factor']

    # columns that will be changed
    to_change = [ 'Nothing in particular', 'Lose weight',
           'Gain weight', 'Stay the same weight', 'Obesity indicator', 'Obesity factor']
    # performing melting
    df_concat = pd.melt(df_concat, id_vars=no_change, value_vars=to_change,
            var_name='Goal', value_name='Goal result')
    
    return df_concat

df_concat = goal(df_concat)
```

2. Education: remove words degree or higher.

```python
# get the titles
def replace_words(df_concat):
    
    df_concat['Education'].apply(lambda x: x.split()[0])
    
    return df_concat

df_concat = replace_words(df_concat)
```

3. Points: data in brakets must be deleted.

```python
# get data in brakets
def get_data_in_brakets(df_concat):
    
    # replace digits in () with empty
    df_concat['Points'] = df_concat['Points'].str.replace('\(\d+\)', '')
    
    # get the first number for those that were not converted by the previous function
    df_concat['Points'] = df_concat['Points'].apply(lambda x: x.split()[0])
    
    return df_concat

df_concat = get_data_in_brakets(df_concat)
```

4. Income must have only Low, middle and High words, create a separate column for the salary values.

```python
# get the first number for those that were not converted by the previous function

def get_income_values(df_concat):
    
    # get the first value from income
    df_concat['Income_values'] = df_concat['Income'].apply(lambda x: x.split()[0])
    
    return df_concat

df_concat = get_income_values(df_concat)


# get salary values from income
def get_salary(df_concat):
    
    # get the second value after splitting and replace () with empty space
    df_concat['Salary'] = df_concat['Income'].apply(lambda x: x.split()[1]).str.replace('(','').str.replace(')', '')
    
    return df_concat

df_concat = get_salary(df_concat)
```

5. Remove values in brakets from income amount, L, food factor and goal.

```python
# get data in brakets for the columns mentioned
def get_data_in_brakets(df_concat):
    
    columns_to_modify = ['Income_amount', 'Living points', 'Food factor', 'Goal result']
    for col in columns_to_modify:
        
        # convert columns to string type
        df_concat[col] = df_concat[col].astype(str)
    
        # replace digits in () with empty
        df_concat[col] = df_concat[col].str.replace('\(\d+\)', '')

        # get the first number for those that were not converted by the previous function
        df_concat[col] = df_concat[col].apply(lambda x: x.split()[0])
    
    return df_concat

df_concat = get_data_in_brakets(df_concat)
```
